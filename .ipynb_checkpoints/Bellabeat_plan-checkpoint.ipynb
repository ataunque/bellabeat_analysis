# Sleep:
Description: Data from each tracked sleep event.

Notes: Sleep durations are either specified by Fitbit wearer (interacting with the device or
Fitbit.com profile), or are automatically detected on supported models (Charge, Alta, Alta HR
Charge HR, Flex, Blaze, Charge 2, Flex 2, Ionic, and Surge). Sleep Stages are supported by
the Alta HR, Charge 2, Blaze, and Ionic. All other devices support the Classic sleep algorithm.

```python
#Daily Sleep Data. TotalSleepRecords: Number of recorded sleep periods for that day. Includes naps > 60min
#Total minutes spent in bed, including asleep,restless, and awake, that occurred during a defined sleep record.
df1 = pd.read_csv('Fitabase Data/Sleep/sleepDay_merged.csv')
display(df1.head())
print(f"Number of Members:{df1['Id'].unique().size}")
```

```python
# Minute SLeep Log. Value indicating the sleep state. 1 = asleep, 2 = restless, 3 = awake
df2 = pd.read_csv('Fitabase Data/Sleep/minuteSleep_merged.csv')
display(df2.head())
print(f"Number of Members:{df2['Id'].unique().size}")
```
